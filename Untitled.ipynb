```python

corpus=[u"我 来到 北京 清华大学",#第一类文本切词后的结果，词之间以空格隔开
        u"他 来到 了 网易 杭研 大厦",#第二类文本的切词结果
        u"小明 硕士 毕业 与 中国 科学院 小明",#第三类文本的切词结果
        u"我 爱 北京 天安门 天安门"]#第四类文本的切词结果
```

```python
v = vectorizer.fit_transform(a)
```

```python
f = open('zeros.txt', 'wb')
import random
i = 0
while i<500:
    j = 10 * random.random() - 5
    k = 10 * random.random() - 5
    c = j * k
    if -0.5 < c < 0.5:
        f.writelines(' '.join([str(j), str(k)]) + '\r\n')
        i += 1
f.close()
```

```python
f = open('ones.txt', 'wb')
import random
i = 0
while i<500:
    j = 10 * random.random() - 5
    k = 10 * random.random() - 5
    c = j * k
    result = ''
    if 0.5 < c < 1.5:
        result += str(j) + ' ' + str(k) + '\r\n'
        i += 1
    f.writelines(result)
f.close()
```
